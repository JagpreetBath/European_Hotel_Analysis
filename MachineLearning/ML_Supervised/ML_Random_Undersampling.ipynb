# European Hotel Analysis Resampling Techniques

## Random Undersampling

```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter

# Reading our Dataset from our hotel_combined.csv file
data = Path('../ML_Supervised/hotel_combined.csv')
df = pd.read_csv(data)
df.head()
```

```python
# Creating a temporary empty column called Review Class at the end of our Dataset
df["Review_Class"] = ""
df
```

```python
# From our table, determining each column's Datatype
df.dtypes
```

```python
# Labelling Reviewer_Scores either Positive or Negative
# For Scores <= 4.9, changing the value in 'review'
df.loc[df["Reviewer_Score"] <= 4.9, "Review_Class"] = "negative"
df.loc[df["Reviewer_Score"] >= 5.0, "Review_Class"] = "positive"
df.head()
```

```python
# Using the LabelEncoder to turn the labels in the Review_Class column into numbers.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = df.copy()
df2['Review_Class'] = le.fit_transform(df2['Review_Class'])
```

```python
df2.head()
```

```python
# Using the LabelEncoder to turn the labels in the Country column into numbers.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df3 = df2.copy()
df3['Country'] = le.fit_transform(df3['Country'])
```

```python
df3.head()
```

```python
# Checking the balance of our target values.  There is a significant inbalance in the direction of the positive reviews.
# This justifies the need for Resampling Machine Learning methods.
df3['Review_Class'].value_counts()
```

```python
# Defining our target variable ('Review_Class') and excluding unecessary columns in our Machine Learning Model.
x_cols = [i for i in df.columns if i not in ('Review_Id', 'Hotel_Address', 'Additional_Number_of_Scoring',
                                             'Review_Date', 'Hotel_Name', 'Reviewer_Nationality', 'Negative_Review', 
                                             'Total_Number_of_Reviews', 'Total_Number_of_Reviews_Reviewer_Has_Given',
                                             'Reviewer_Score', 'Tags', 'days_since_review', 'City',
                                             'Positive_Review', 'Review_Class')]
X = df3[x_cols]
y = df3['Review_Class']
```

```python
# Splitting our data into testing and training sets.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Resampling the training data with the RandomUndersampler
from imblearn.under_sampling import RandomUnderSampler
ros = RandomUnderSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
from sklearn import utils
print(utils.multiclass.type_of_target(y_train))
```

```python
# Checking to see the variables in X_resampled (verifying whether our target and unecessary columns were properly identified)
X_resampled
```

```python
# Checking to see the variables in y_resampled (verifying whether our target and unecessary columns were properly identified)
y_resampled
```

```python
# Training the Logistic Regression model using the resampled data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Displaying the confusion matrix
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculating the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

```python
# Printing the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```
