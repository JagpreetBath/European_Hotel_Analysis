```python
# importing dependencies
import pandas as pd
import zipfile as zf
```

```python
# importing Table 1 and converting to a df
zip_file = zf.ZipFile('../DataFiles/TransformedData/tables/hotel_reviews.zip') 
hotel_reviews = pd.read_csv(zip_file.open('hotel_review.csv'))

# check the df
hotel_reviews
```

```python
# importing Table 2 and converting to a df
zip_file = zf.ZipFile('../DataFiles/TransformedData/tables/hotel_location_city_country.zip') 
hotel_location_city_country = pd.read_csv(zip_file.open('hotel_location_city_country.csv'))

# since Hotel_Name occurs in both Tables, will drop it from Table 2
hotel_location_city_country = hotel_location_city_country.drop(columns=['Hotel_Name'])

# check the df
hotel_location_city_country
```

```python
# merging the two tables on the column 'Hotel_Address'
hotel_combined = hotel_reviews.merge(hotel_location_city_country, on='Hotel_Address')

hotel_combined
```

```python
# setting the 'Review_Id' as the index
hotel_combined.set_index('Review_Id',inplace=True)
hotel_combined
```

# Exporting CSV as a Zip file

```python
# Exporting hotel_combined.csv as a Zip file
compression_opts = dict(method='zip', archive_name='hotel_and_location_combined_all_rows.csv')  
hotel_combined.to_csv('../DataFiles/TransformedData/tables/hotel_and_location_combined_all_rows.zip', index=True, compression=compression_opts)  
```

# Exporting JSON as a ZIP file

```python
# Exporting hotel_combined.json as a Zip file
compression_opts = dict(method='zip', archive_name='all_rows_JSON.json')  
hotel_combined.to_json('../DataFiles/TransformedData/JSON/all_rows_JSON.zip', index=True, compression=compression_opts) 
```

# Reading the JSON to make sure it was exported correctly in the last step

```python
# Reading the JSON to make sure it was exported correctly
zip_file = zf.ZipFile('../DataFiles/TransformedData/JSON/all_rows_JSON.zip') 
df = pd.read_json(zip_file.open('all_rows_JSON.json'))

# check the df
df
```
