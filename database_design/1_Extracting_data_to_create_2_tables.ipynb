```python
# importing dependencies
import pandas as pd
import zipfile as zf
```

```python
# importing file and converting to a df
zip_file = zf.ZipFile('../Resources/archive.zip') 
hotel_df = pd.read_csv(zip_file.open('Hotel_Reviews.csv'))
hotel_df.index.name='Id'

# check the df
hotel_df 
```

```python
# check for null/na values
hotel_df.isnull().sum() 
```

```python
# drop rows that are na/null
hotel_df = hotel_df.dropna() 
hotel_df 
```

```python
# check for null again
hotel_df.isnull().sum() 
```

# Breaking the big table into 2 tables: 1st with hotel reviews, 2nd with hotel locations

```python
## Table 1: Hotel reviews only - no location data

# hotel reviws without lat and lng
hotel_reviews_without_location = hotel_df.drop(columns=['lat','lng'])
hotel_reviews_without_location 
```

```python
## Table 2: DF with locations (no reviews)

# creating a subset of the dataframe with unique values on the column 'Hotel_Address'
unique_address_df = hotel_df.drop_duplicates(subset=['Hotel_Address'])

# droping all columns excpet Hotel Address, Name, lat and lng
hotel_location = unique_address_df.drop(columns =['Additional_Number_of_Scoring', 'Review_Date', 'Average_Score', 'Reviewer_Nationality', 'Negative_Review', 'Review_Total_Negative_Word_Counts', 'Total_Number_of_Reviews','Positive_Review', 'Review_Total_Positive_Word_Counts', 'Total_Number_of_Reviews_Reviewer_Has_Given', 'Reviewer_Score', 'Tags', 'days_since_review' ])
hotel_location
```

# Exporting as csv files

```python
# exporting as csv files
hotel_reviews_without_location.to_csv('tables/hotel_reviews_for_project.csv', index=True)
hotel_location.to_csv('tables/hotel_location_for_project.csv', index=False)
```
